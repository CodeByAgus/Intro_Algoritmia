```python
# Ejercicio 1: Ingresar dos números enteros e indicar si son iguales o distintos.
a = int(input('Ingrese el primer número entero: '))
b = int(input('Ingrese el segundo número entero: '))
if a == b:
    print('Iguales')
else:
    print('Distintos')
```

```python
# Ejercicio 2: Leer un número entero e imprimir si es par o impar.
n = int(input('Ingrese un número entero: '))
if n % 2 == 0:
    print('Par')
else:
    print('Impar')
```

```python
# Ejercicio 3: Pedir número de mes e imprimir su nombre; validar rango.
mes = int(input('Ingrese el número de mes (1-12): '))
meses = ['enero','febrero','marzo','abril','mayo','junio','julio','agosto','septiembre','octubre','noviembre','diciembre']
if 1 <= mes <= 12:
    print(meses[mes-1])
else:
    print('Mes inválido')
```

```python
# Ejercicio 4: Ingresar notas de dos parciales e indicar estado (promocionó/aprobó/recuperar).
n1 = float(input('Nota primer parcial (0-10): '))
n2 = float(input('Nota segundo parcial (0-10): '))
if not (0 <= n1 <= 10) or not (0 <= n2 <= 10):
    print('Error: nota fuera del rango 0-10')
else:
    if n1 >= 7 and n2 >= 7:
        print('Promocionó')
    elif n1 >= 4 and n2 >= 4:
        print('Aprobó')
    else:
        print('Debe recuperar')
```

```python
# Ejercicio 5: Calcular costo de un libro según número de páginas.
paginas = int(input('Ingrese el número de páginas del libro: '))
costo_base = 50000
costo_por_pagina = 2000.20
costo = costo_base + paginas * costo_por_pagina
if paginas > 300:
    costo += 20000  # encuadernación en tela
if paginas > 600:
    costo += 33600  # procedimiento especial
print(f'Costo total: ${costo:.2f}')
```

```python
# Ejercicio 6: Calcular precio de viaje según km con tarifa y viaje mínimo.
km = float(input('Ingrese distancia en km: '))
if km < 0:
    print('Distancia inválida')
else:
    if km < 10:
        precio = km * 2000
    else:
        precio = km * 1500
    precio = max(precio, 5000)
    print(f'Precio del viaje: ${precio:.2f}')
```

```python
# Ejercicio 7: Leer un año e indicar si es bisiesto.
anio = int(input('Ingrese un año (por ejemplo 2024): '))
is_bisiesto = False
if anio % 4 == 0:
    is_bisiesto = True
    if anio % 100 == 0 and anio % 400 != 0:
        is_bisiesto = False
print('Bisiesto' if is_bisiesto else 'No es bisiesto')
```
