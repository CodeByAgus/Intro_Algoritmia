```python
# Ejercicio 4: Leer 10 números enteros e imprimir el promedio, el mayor y en qué orden fue ingresado el mayor valor.
# Si se ingresó más de una vez debe informar el primer ingreso.

v = []
for i in range(10):
    n = int(input("Ingrese un número entero: "))
    v.append(n)

promedio = sum(v) / len(v)
mayor = max(v)
orden_mayor = v.index(mayor) + 1

print(f"Promedio: {promedio}")
print(f"Mayor: {mayor}")
print(f"Orden de ingreso del mayor: {orden_mayor}")
```

```python
# Ejercicio 6: Ingresar números, hasta que la suma de los números pares supere 100. 
# Mostrar cuántos números en total se ingresaron.
suma_pares = 0
contador = 0

while suma_pares <= 100:
    numero = int(input("Ingrese número: "))
    contador += 1
    if numero % 2 == 0:
        suma_pares += numero

print(f"Se ingresaron {contador} números. Suma de pares = {suma_pares}")


```

```python
# Ejercicio 7: Ingresar importes de ventas hasta indicar -1; 
# informar monto total vendido y cuántas ventas se realizaron.
# El importe de cada venta debe ser un valor positivo.
total_ventas = 0
monto_total = 0
num = 0

while num != -1:
    num = int(input("Ingrese venta (o -1 para terminar): "))
    if num != -1:
        monto_total += num
        total_ventas += 1

print(f"Monto total vendido: {monto_total}")
print(f"Cantidad de ventas: {total_ventas}")

```
