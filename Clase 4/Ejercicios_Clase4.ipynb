```python
# Ejercicio 4: Leer 10 números enteros e imprimir el promedio, el mayor y en qué orden fue ingresado el mayor valor.
# Si se ingresó más de una vez debe informar el primer ingreso.

v = []
for i in range(10):
    n = int(input("Ingrese un número entero: "))
    v.append(n)

promedio = sum(v) / len(v)
mayor = max(v)
orden_mayor = v.index(mayor) + 1

print(f"Promedio: {promedio}")
print(f"Mayor: {mayor}")
print(f"Orden de ingreso del mayor: {orden_mayor}")
```
