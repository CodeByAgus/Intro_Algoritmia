```python
# Ejercicio 1:
# a) Mostrar el mensaje 'Hola Mundo'.
# b) Ingresar el nombre del usuario y saludarlo (ej: 'Hola Juan').
# c) Ingresar dos números y mostrar la suma y la diferencia.
# d) Ingresar tres números y mostrar la suma y el promedio.
# e) Ingresar el monto de una factura y calcular el IVA (21%).

print("Hola Mundo") #imprime el mensaje "Hola Mundo"
nombre = input("Ingrese su nombre: ") #solicita al usuario que ingrese su nombre
print("Hola " + nombre) #saluda al usuario con su nombre

num1 = int(input("Ingrese el primer número: ")) #solicita al usuario que ingrese el primer número 
num2 = int(input("Ingrese el segundo número: ")) #solicita al usuario que ingrese el segundo número 
suma = num1 + num2 #calcula la suma de los dos números
diferencia = num1 - num2 #calcula la diferencia de los dos números              
print(f"La suma es: {suma}") #imprime la suma de los dos números
print(f"La diferencia es: {diferencia}") #imprime la diferencia de los dos números
num3 = int(input("Ingrese el tercer número: ")) #solicita al usuario que ingrese el tercer número (int)
suma_total = num1 + num2 + num3 #calcula la suma de los tres números
promedio = suma_total / 3 #calcula el promedio de los tres números
print(f"La suma total es: {suma_total}") #imprime la suma total de los tres números
print(f"El promedio es: {promedio}") #imprime el promedio de los tres números

monto_factura = float(input("Ingrese el monto de la factura: ")) #solicita al usuario que ingrese el monto en float
iva = monto_factura * 0.21 #calcula el IVA (21% del monto de la factura)
```

```python
# Ejercicio 5: Calcular la ganancia mensual y a seis meses de una inversión con 2% mensual.
# Entrada: capital. Salidas: ganancia en 1 mes y ganancia en 6 meses (simple/interés compuesto a especificar).
capital = float(input("Ingrese el capital invertido: ")) #solicita ingresar el capital invertido
ganancia_mensual = capital * 0.02 #calcula la ganancia mensual utilizando interés simple
ganancia_seis_meses_simple = ganancia_mensual * 6 #calcula la ganancia a seis meses utilizando interés simple
ganancia_seis_meses_compuesto = round(capital * (1 + 0.02) ** 6 - capital)   #calcula la ganancia a seis meses utilizando interés compuesto
print(f"La ganancia mensual es: {ganancia_mensual}") #imprime la ganancia mensual
print(f"La ganancia a seis meses (interés simple) es: {ganancia_seis_meses_simple}") #imprime la ganancia a seis meses utilizando interés simple
print(f"La ganancia a seis meses (interés compuesto) es: {ganancia_seis_meses_compuesto}") #imprime la ganancia a seis meses utilizando interés
```

```python
# Ejercicio 6: Leer una medida en metros e imprimir en cm, pulgadas, pies y yardas.
# Factores: 1 m = 100 cm; 1 pulgada = 2.54 cm; 1 pie = 12 pulgadas; 1 yarda = 3 pies.

medida_metros = float(input("Ingrese la medida en metros: "))   
medida_cm = medida_metros * 100 #convierte la medida de metros a centímetros
medida_pulgadas = medida_cm / 2.54 #convierte la medida de   centímetros a pulgadas
medida_pies = medida_pulgadas / 12 #convierte la medida de pulgadas a pies
medida_yardas = medida_pies / 3 #convierte la medida de pies a yardas
print(f"La medida en centímetros es: {medida_cm} cm") #imprime la medida en centímetros
print(f"La medida en pulgadas es: {medida_pulgadas} pulgadas") #imprime la medida en pulgadas
print(f"La medida en pies es: {medida_pies} pies") #imprime la medida en pies
print(f"La medida en yardas es: {medida_yardas} yardas") #imprime la medida en yardas
```

```python
# Ejercicio 9: Leer un período en segundos e imprimir en días, horas, minutos y segundos.
# Ejemplo: 200000 s = 2 días, 7 horas, 33 minutos, 20 segundos.

segundos = int(input("Ingrese periodo en segundos"))  # solicita al usuario ingresar el período en segundos
dias = segundos // 86400  # calcula los días dividiendo segundos entre 86400 (segundos en un día)
restos = segundos % 86400  # calcula los segundos restantes después de extraer los días
horas = restos // 3600  # calcula las horas dividiendo los segundos restantes entre 3600 (segundos en una hora)
restos %= 3600  # actualiza los segundos restantes después de extraer las horas
minutos = restos // 60  # calcula los minutos dividiendo los segundos restantes entre 60 (segundos en un minuto)
segundos_restantes = restos % 60  # calcula los segundos finales restantes

print(f"{dias} días, {horas} horas, {minutos} minutos, {segundos_restantes} segundos")
```

```python
# Ejercicio 10: Convertir un número binario de 4 cifras a decimal.
# Se ingresa como un entero de cuatro dígitos; usar sumatoria de dígitos * 2^posicion.
binario = int(input("Ingrese numero de 4 digitos"))

# Extraer dígitos usando operaciones básicas (división entera y módulo)
d1 = binario // 1000  # primer dígito (izquierda) 1xxx//mil = cantidad de miles 0xxx//mil = 0
d2 = (binario // 100) % 10  # segundo dígito x1xx // redondeado % 10 (0 o 1)
d3 = (binario // 10) % 10  # tercer dígito 
d4 = binario % 10  # cuarto dígito (derecha)

print (d1)
print(d2)
print(d3)
print(d4)


# Calcular decimal
decimal = d1 * 8 + d2 * 4 + d3 * 2 + d4 * 1

print(f"El número binario {binario} en decimal es: {decimal}")


```
